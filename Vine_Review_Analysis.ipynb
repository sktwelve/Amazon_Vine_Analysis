```python
import pandas as pd
```

```python
vine_df = pd.read_csv("./vine_table.csv")

vine_df
```

```python
vine_df_20 = vine_df.loc[vine_df["total_votes"] >= 20]

vine_df_20
```

```python
vine_df_half = vine_df_20.loc[vine_df_20["helpful_votes"]/vine_df_20["total_votes"] >= .5]

vine_df_half
```

```python
vine_df_y = vine_df_half.loc[vine_df_half["vine"] == "Y"]

vine_df_y
```

```python
vine_df_n = vine_df_half.loc[vine_df_half["vine"] == "N"]

vine_df_n
```

```python
# Vine 5 Star Reviews
y_5star = vine_df_y.loc[vine_df_y["star_rating"] == 5]["star_rating"].count()
y_5star 
```

```python
# Non Vine 5 Star Reviews
n_5star = vine_df_n.loc[vine_df_n["star_rating"] == 5]["star_rating"].count()
n_5star 
```

```python
# Vine Reviews
y_reviews = vine_df_y["star_rating"].count()
y_reviews
```

```python
# Non Vine Reviews
n_reviews = vine_df_n["star_rating"].count()
n_reviews
```

```python
# Percentage of 5 Star Vine Reviews 
y_percent = (y_5star/y_reviews)*100
y_percent
```

```python
# Percentage of 5 Star Non Vine Reviews 
n_percent = (n_5star/n_reviews)*100
n_percent
```
